## Arrays

**Arrays have a lot in common with lists:**


*   Both arrays and lists store elements in a sequential manner, meaning you can access elements by their position or index in the collection.
*   Most of the operations have a similar syntax as we will see

**But they also have some differences:**


*   All elements in an array are of the same type while lists' elements can be of different types. This makes arrays more memory-efficient and faster for numerical and scientific computing operations.
*   Lists are built-in data structures while arrays are provided by external libraries like NumPy.

***So how do we decide whether we want to use lists or arrays ?***

Lists are more versatile and suitable for general purposes, while arrays are specialized data structures optimized for numerical and scientific computing tasks.

## Stacks

Picture a stack of books on a desk. You put the first one down, then stack the next one on top of that, then a third.

What order do you take them off?

That's how a stack works in Python, too. The last thing you put on the stack is the first thing you take off of it.

This is called "LIFO", or "last in, first out." The last book you stacked on the pile is the first one you grab to take off the pile.

```python
# Initializing a stack using a list 
stack = []

# Adding elements to the stack
stack.append('a')
stack.append('b')
stack.append('c')

print(f"Initial stack: {stack}")

# Removing elements from a stack
print(f"\nstack.pop() gives us: {stack.pop()}")

print(f"\nStack after element is popped: {stack}")
```

## Queues

Queues are sort of the opposite of stacks. Picture a line in front of a food truck. The first person in line is the first person to get served, and the second person in line is the second person to get served, etc.

In a queue, the first item placed in the queue is the first item to be removed from it.

This is called "FIFO", or "first in, first out."

```python
# Initialize the queue using a list, again
queue = []

# Adding elements to the queue
queue.append('a')
queue.append('b')
queue.append('c')

print(f"Initial queue: {queue}")

# Removing elements from the queue
print(f'\nqueue.pop(0) gets us: {queue.pop(0)}')

print(f"\nQueue after elements are dequeued: {queue}")
```

Note the difference here:

```python
stack.pop()
queue.pop(0)
```

The function ".pop()" with empty parentheses will always remove the last item in the list, but when you want a list to *act* like a queue, you have to specify that the item you want to remove can be found in spot "0", which is why you put the "0" between the parentheses when you pop from a queue.

## Bonus after class: Errors in Python



Python, like any language, has its own way of communicating when something goes wrong in your code. These messages, called "errors," are your computer's way of telling you what's gone awry. Let's look at three common types of Python errors:



### Syntax error

A syntax error occurs when you violate the rules of Python's grammar. Here's an example:

```python
if x = 5:
    print("Hello, World!")
```

In this code, we're using the assignment operator `=` instead of the equality operator `==` inside the `if` statement. Python will raise a syntax error because it expects a valid condition after `if`.

### Runtime error

A runtime error occurs during program execution when something unexpected happens. See below for an example:

```python
num = 10
den = 0
result = num / den
```

In this code, we're trying to divide `num` by `den`, where `den` is zero. Division by zero is mathematically impossible and will cause a runtime error, more specifically a "ZeroDivisionError".

Another example of common runtime errors you are likely to encounter is running a program with misspelled or incorrectly capitalized variable and function names.

### Logical error

Last but not least, logical errors are common errors that don't raise an error but don't return the expected result. In the example below, the code executes but will only produce an actual average when the list provided is of length 2:

```python
def calculate_average(numbers):
    total = 0
    for number in numbers:
        total += number
    average = total / 2  # Logical error: Dividing by 2 instead of len(numbers)
    return average

numbers = [10, 20, 30, 40]
avg = calculate_average(numbers)
print("Average:", avg)
```

## Try ... except ...

Not all errors can be prevented but they can be managed and `try` ... `except` ... is a very common way to handle errors when we can't prevent them.

```python
try:
    # Code that might raise an error
except ErrorType:
    # Handle the error
```

If we go back to our RuntimeError example, we can handle it with the following `try...except`:

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Oops! You can't divide by zero.")
```

```python
# You can also handle multiple error types

try:
    num = int(input("Enter a number: "))
    result = 10 / num
except (ValueError, ZeroDivisionError):
    print("Invalid input or division by zero.")
```

This is even more valuable when you have several `except` to handle the different error types to inform the user of what happened. The example below refers to error handling when doing web scraping:

```python
import requests
from bs4 import BeautifulSoup

# Define the URL you want to scrape
url = "https://www.nytimes.com/" # As we are New Yorkers, let's look at the NY Times live news

try:
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    response.raise_for_status()

    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract data
    data = soup.find('div', class_='css-100xb1c e1ppw5w20').text.strip() # You can look at the code behind the page with a right click then inspect

except requests.exceptions.RequestException as req_err:
    print(f"Request error: {req_err}")
except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error: {http_err}")
except requests.exceptions.ConnectionError as conn_err:
    print(f"Connection error: {conn_err}")
except requests.exceptions.Timeout as timeout_err:
    print(f"Timeout error: {timeout_err}")
except requests.exceptions.TooManyRedirects as redirect_err:
    print(f"Too many redirects error: {redirect_err}")
except requests.exceptions.RequestException as req_err:
    print(f"Request error: {req_err}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
else:
    print(f"Data: {data}")

finally:
    if 'response' in locals():
        response.close()
```

```python
"""
GWC-CU
Week 2 Exercise 2 - Hangman Game
"""

import random
import string

# Open the dictionary.txt file and choose a random word
file = open("dictionary.txt", "r") # This might not work on Google Colab but will work if you're using the Replit we provided you as the dictionary i
lines = file.readlines()

random_word = lines[random.randint(0, len(lines) - 1)].strip().lower()
file.close()

# Create a function that displays the word with guessed letters that will be called at each round
def display_word(word, guessed_letters):
    displayed = ""
    for letter in word:
        if letter in guessed_letters:
            displayed += letter + " "
        else:
            displayed += "_ "
    return displayed

# And here is the main Hangman game function that we call to start the game
def play_hangman():
    max_attempts = 5  # Maximum incorrect guesses allowed
    guessed_letters = []
    attempts = 0

    print("Welcome to Hangman!")
    print(display_word(random_word, guessed_letters))

    while True:
        guess = input("Guess a letter: ").lower()

        if guess in guessed_letters:
            print("You already guessed that letter.")
            continue

        guessed_letters.append(guess)

        if guess in random_word:
            print("Correct guess!")
        else:
            print("Incorrect guess.")
            attempts += 1

        print(display_word(random_word, guessed_letters))

        if set(random_word).issubset(set(guessed_letters)):
            print(f"Congratulations! You've guessed the word: {random_word}")
            break

        # We don't allow the user to do more than max_attempts wrong guesses
        if attempts >= max_attempts:
            print(f"Sorry, you've reached the maximum allowed attempts. The word was: {random_word}")
            break

# Start the game
play_hangman()
```
